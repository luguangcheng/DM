```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
goods_train = pd.read_table('./vp_data/vp_data/goods_train.txt',header=None)   #2004581 spu  6955 brand  49 cat
user_action_train = pd.read_table('./vp_data/vp_data/user_action_train.txt',header=None)# 56172886 click 196030 user  1920715 spu  48 cat 6497 brand  83 date 
user_action_test_items = pd.read_table('./vp_data/vp_data/user_action_test_items.txt',header=None)
```

```python
goods_train.columns=['spu_id','brand_id','cat_id']
user_action_train.columns = ['user_id','spu_id','action_type','date']
user_action_test_items,columns = ['user_id','spu_id']
```

```python
#去除异常特征值
user_action_train = user_action_train.groupby(['user_id','spu_id','date'],as_index = False).sum()
```

```python
user_goods_action_train = pd.merge(user_action_train,goods_train, how='left', on='spu_id')
user_goods_action_train
```

```python
#将日期离散化（0-83）
date_set = user_goods_action_train['date'].drop_duplicates().sort_values().values
date ={'date':date_set,'date_num':range(83)}
date_df = pd.DataFrame(date)
user_goods_action_train = pd.merge(user_goods_action_train,date_df, how='left', on='date')
user_goods_action_train
```

```python
#将数据按时间排序
user_goods_action_train_sort = user_goods_action_train.sort_values(by = 'date_num')
user_goods_action_train_sort
```

```python
#用于生成与用户相关的特征
def get_user_feature(df, start_date, end_date):
    df = df[(df['date_num'] >= start_date) & (df['date_num'] < end_date)]
    #用户购买量——buy
    user_feature = df[['user_id','action_type']].groupby('user_id',as_index=False).sum()
    user_feature.rename(columns={'action_type':'buy'}, inplace = True)
    #用户点击量——user_click
    click = df.groupby('user_id',as_index=False).size()
    click.index = range(len(click))
    user_feature['user_click'] = click
    #用户购买率——buy_click_ratio
    user_feature['buy_click_ratio'] = user_feature['buy']/user_feature['user_click']
    #计算用户最早与最后交互时间与预测时间的间隔——user_first_action_date、user_last_action_date
    df_first = df[['user_id','date_num']].drop_duplicates(['user_id'])
    df_last = df[['user_id','date_num']].drop_duplicates(['user_id'], keep='last')    
    df_first['user_first_action_date'] = end_date - df_first.loc[:,'date_num']
    df_last['user_last_action_date'] = end_date - df_last['date_num']
    del df_first['date_num']
    del df_last['date_num']
    user_feature = pd.merge(user_feature,df_first,how='left', on='user_id')
    user_feature = pd.merge(user_feature,df_last,how='left', on='user_id')
    #计算隔天购买数——buy_getian
    df_buy = df[df['action_type']==1]
    df_buy = df_buy[['user_id','spu_id']]
    df_buy_act = pd.merge(df_buy, df, how='left', on=['user_id','spu_id'])
    df_buy_act = df_buy_act[['user_id','spu_id','action_type']]
    df_buy_act = df_buy_act.drop_duplicates(['user_id','spu_id'])
    df_buy_act = df_buy_act[['user_id','action_type']].groupby(['user_id'],as_index=False).sum()
    user_feature = pd.merge(user_feature,df_buy_act,how = 'left',on='user_id')
    user_feature = user_feature.fillna(0)
    user_feature['buy_getian']= user_feature['buy']- user_feature['action_type']
    #隔天购买率——buy_getian_ratio
    user_feature['buy_getian_ratio'] =  user_feature['buy_getian']/user_feature['buy']
    del user_feature['action_type']
    user_feature = user_feature.fillna(0)
    return user_feature
```

```python
#用于生成与商品相关的特征
def get_spu_feature(df, start_date, end_date):
    df = df[(df['date_num'] >= start_date) & (df['date_num'] < end_date)]
    #商品购买量——sell
    spu_feature = df[['spu_id','action_type']].groupby('spu_id',as_index=False).sum()
    spu_feature.rename(columns={'action_type':'sell'}, inplace = True)
    #商品点击量——spu_click
    spu_click = df.groupby('spu_id',as_index=False).size()
    spu_click.index = range(len(spu_click))
    spu_feature['spu_click'] = spu_click
    #商品购买率——sell_click_ratio
    spu_feature['sell_click_ratio'] = spu_feature['sell']/spu_feature['spu_click']
    #计算商品最早与最后交互时间与预测时间的间隔——spu_first_action_date、spu_last_action_date
    df_first = df[['spu_id','date_num']].drop_duplicates(['spu_id'])
    df_last = df[['spu_id','date_num']].drop_duplicates(['spu_id'], keep='last')    
    df_first['spu_first_action_date'] = end_date - df_first.loc[:,'date_num']
    df_last['spu_last_action_date'] = end_date - df_last['date_num']
    del df_first['date_num']
    del df_last['date_num']
    spu_feature = pd.merge(spu_feature,df_first,how='left', on='spu_id')
    spu_feature = pd.merge(spu_feature,df_last,how='left', on='spu_id')
    #计算商品隔天被购买数——spu_sell_getian
    df_buy = df[df['action_type']==1]
    df_buy = df_buy[['user_id','spu_id']]
    df_buy_act = pd.merge(df_buy, df, how='left', on=['user_id','spu_id'])
    df_buy_act = df_buy_act[['user_id','spu_id','action_type']]
    df_buy_act = df_buy_act.drop_duplicates(['user_id','spu_id'])
    df_buy_act = df_buy_act[['spu_id','action_type']].groupby(['spu_id'],as_index=False).sum()
    spu_feature = pd.merge(spu_feature,df_buy_act,how = 'left',on='spu_id')
    spu_feature = spu_feature.fillna(0)
    spu_feature['spu_sell_getian']= spu_feature['sell']- spu_feature['action_type']
    #商品隔天被购买率——spu_sell_getian_ratio
    spu_feature['spu_sell_getian_ratio'] =  spu_feature['spu_sell_getian']/spu_feature['sell']
    del spu_feature['action_type']
    spu_feature = spu_feature.fillna(0)
    #对商品的类别进行one-hot编码
    spu_feature = pd.merge(spu_feature,goods_train,how='left',on='spu_id')
    cat_dummy = pd.get_dummies(spu_feature['cat_id'],prefix='cat_id')
    spu_feature = pd.concat([spu_feature,cat_dummy],axis=1)
    del spu_feature['cat_id']
    del spu_feature['brand_id']
    return spu_feature
```

```python
#用于生成与类别（cate）相关的特征
def get_cat_feature(df, start_date, end_date):
    df = df[(df['date_num'] >= start_date) & (df['date_num'] < end_date)]
    #类别购买量——cat_sell
    cat_feature = df[['cat_id','action_type']].groupby('cat_id',as_index=False).sum()
    cat_feature.rename(columns={'action_type':'cat_sell'}, inplace = True)
    #类别点击量——cat_click
    cat_click = df.groupby('cat_id',as_index=False).size()
    cat_click.index = range(len(cat_click))
    cat_feature['cat_click'] = cat_click
    #类别购买率——cat_sell_click_ratio
    cat_feature['cat_sell_click_ratio'] = cat_feature['cat_sell']/cat_feature['cat_click']
    #计算品类最早与最后交互时间与预测时间的间隔——cat_first_action_date、cat_last_action_date
    df_first = df[['cat_id','date_num']].drop_duplicates(['cat_id'])
    df_last = df[['cat_id','date_num']].drop_duplicates(['cat_id'], keep='last')    
    df_first['cat_first_action_date'] = end_date - df_first.loc[:,'date_num']
    df_last['cat_last_action_date'] = end_date - df_last['date_num']
    del df_first['date_num']
    del df_last['date_num']
    cat_feature = pd.merge(cat_feature,df_first,how='left', on='cat_id')
    cat_feature = pd.merge(cat_feature,df_last,how='left', on='cat_id')
    ##计算类别隔天被购买数——cat_sell_getian
    df_buy = df[df['action_type']==1]
    df_buy = df_buy[['user_id','spu_id']]
    df_buy_act = pd.merge(df_buy, df, how='left', on=['user_id','spu_id'])
    df_buy_act = df_buy_act[['user_id','spu_id','cat_id','action_type']]
    df_buy_act = df_buy_act.drop_duplicates(['user_id','spu_id'])
    df_buy_act = df_buy_act[['cat_id','action_type']].groupby(['cat_id'],as_index=False).sum()
    cat_feature = pd.merge(cat_feature,df_buy_act,how = 'left',on='cat_id')
    cat_feature = cat_feature.fillna(0)
    cat_feature['cat_sell_getian']= cat_feature['cat_sell']- cat_feature['action_type']
    #类别隔天购买率——cat_sell_getian_ratio
    cat_feature['cat_sell_getian_ratio'] =  cat_feature['cat_sell_getian']/cat_feature['cat_sell']
    del cat_feature['action_type']
    cat_feature = cat_feature.fillna(0)
    return cat_feature
```
