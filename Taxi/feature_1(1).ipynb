```python
# 叫车时间-月，周，小时 # 后续 dummy
allData['pickup_datetime'] = pd.to_datetime(allData.pickup_datetime)
allData['pickup_month'] = allData['pickup_datetime'].dt.month - 1
allData['pickup_weekday'] = allData['pickup_datetime'].dt.weekday
allData['pickup_hour'] = allData['pickup_datetime'].dt.hour
```

```python
# 经纬度离散化 # 后续 dummy
n = 10;m = 10;
allData['pickup_longitude_qcut'] = pd.qcut(allData['pickup_longitude'],n,labels=range(n))
allData['pickup_latitude_qcut'] = pd.qcut(allData['pickup_latitude'],m,labels=range(m))
allData['dropoff_longitude_qcut'] = pd.qcut(allData['dropoff_longitude'],n,labels=range(n))
allData['dropoff_latitude_qcut'] = pd.qcut(allData['dropoff_latitude'],m,labels=range(m))
```

```python
# 距离特征，街区距离与直线距离
def Lon_Lat_distance(lon1,lat1,lon2,lat2):
    deltaLon = np.abs(lon1 - lon2)
    deltaLat = np.abs(lat1 - lat2)
    EARTH_RADIUS = 6371
    disLon,disLat = map(lambda x:np.pi*x*EARTH_RADIUS/180,(deltaLon,deltaLat))
    return disLon, disLat

def calcu_distance(lon1,lat1,lon2,lat2):
    disLon,disLat = Lon_Lat_distance(lon1,lat1,lon2,lat2)
    mht_distance = disLon + disLat
    line_distance = np.sqrt(disLon**2 + disLat**2)
    return mht_distance, line_distance

# 两种距离特征（+2）
allData['mht_distance'],allData['line_distance'] = calcu_distance(allData['pickup_longitude'],allData['pickup_latitude'],
                                                                  allData['dropoff_longitude'],allData['dropoff_latitude'])
```

```python
# 方向特征，连续，离散
allData['longitude_delta'] = allData['dropoff_longitude'] - allData['pickup_longitude']
allData['latitude_delta'] = allData['dropoff_latitude'] - allData['pickup_latitude']

allData['longitude_dct'] = (allData.longitude_delta > 0)*1
allData['latitude_dct'] = (allData.latitude_delta > 0)*1
```

```python
# 速度特征, 后续删除
allData['mht_speed'] = 1000 * allData['mht_distance'] / allData['trip_duration']
allData['line_speed'] = 1000 * allData['line_distance'] / allData['trip_duration']
```

```python
# mht_region_speed, line_region_speed, region_trip_duration
def fillnan_region_4(a,b,c,d,df,col):
    a1=max(a-1,0);a2=min(a+1,9);
    b1=max(b-1,0);b2=min(b+1,9);
    c1=max(c-1,0);c2=min(c+1,9);
    d1=max(d-1,0);d2=min(d+1,9);
    s1=df[((df['pickup_longitude_qcut']>=a1) & (df['pickup_longitude_qcut']<=a2)) & 
          ((df['pickup_latitude_qcut']>=b1) & (df['pickup_latitude_qcut']<=b2)) & 
          (df['dropoff_longitude_qcut']==c) & (df['dropoff_latitude_qcut']==d)][col].mean()
    s2=df[(df['pickup_longitude_qcut']==a) & (df['pickup_latitude_qcut']==b) & 
          ((df['dropoff_longitude_qcut']>=c1) & (df['dropoff_longitude_qcut']<=c2)) &
          ((df['dropoff_latitude_qcut']>=d1) & (df['dropoff_latitude_qcut']<=d2))][col].mean()
    return (s1+s2)/2.0

def fillnan_region_9(a,b,c,d,df,col):
    a1=max(a-1,0);a2=min(a+1,9);
    b1=max(b-1,0);b2=min(b+1,9);
    c1=max(c-1,0);c2=min(c+1,9);
    d1=max(d-1,0);d2=min(d+1,9);
    s=df[((df['pickup_longitude_qcut']>=a1) & (df['pickup_longitude_qcut']<=a2)) & 
         ((df['pickup_latitude_qcut']>=b1) & (df['pickup_latitude_qcut']<=b2)) & 
         ((df['dropoff_longitude_qcut']>=c1) & (df['dropoff_longitude_qcut']<=c2)) &
         ((df['dropoff_latitude_qcut']>=d1) & (df['dropoff_latitude_qcut']<=d2))][col].mean()
    return s

def fill_region(df,col):
    count = 1
    while(count != 0):
        count = 0
        for i in range(df.shape[0]):
            if df.iloc[i,:].isnull()[4]:
                df.iloc[i,4] = fillnan_region_4(i/1000, i%1000/100, i%100/10, i%10, df,col) # 填充缺失值方式
                if df.iloc[i,:].isnull()[4]:
                    count += 1
        print count
    return df
```

```python
# mht_region_speed_m, line_region_speed_m, region_trip_duration_m
def fillnan_region_mwh_4(m,a,b,c,d,df,col,dat,n):
    m1=max(m-1,0);m2=min(m+1,n)
    a1=max(a-1,0);a2=min(a+1,9);
    b1=max(b-1,0);b2=min(b+1,9);
    c1=max(c-1,0);c2=min(c+1,9);
    d1=max(d-1,0);d2=min(d+1,9);
    s1=df[((df[dat]>=m1) & (df[dat]<=m2)) &
          ((df['pickup_longitude_qcut']>=a1) & (df['pickup_longitude_qcut']<=a2)) & 
          ((df['pickup_latitude_qcut']>=b1) & (df['pickup_latitude_qcut']<=b2)) & 
          (df['dropoff_longitude_qcut']==c) & (df['dropoff_latitude_qcut']==d)][col].mean()
    s2=df[((df[dat]>=m1) & (df[dat]<=m2)) &
          (df['pickup_longitude_qcut']==a) & (df['pickup_latitude_qcut']==b) & 
          ((df['dropoff_longitude_qcut']>=c1) & (df['dropoff_longitude_qcut']<=c2)) &
          ((df['dropoff_latitude_qcut']>=d1) & (df['dropoff_latitude_qcut']<=d2))][col].mean()
    return (s1+s2)/2.0

def fill_region_mwh(df,col):
    if col[-1] == 'm': 
        dat = 'pickup_month'; n = 5
    elif col[-1] == 'w': 
        dat = 'pickup_weekday'; n = 6
    elif col[-1] == 'h': 
        dat = 'pickup_hour'; n = 23
    else:
        print 'error:',col[-1]
        return     
    count = 1
    while(count != 0):
        count = 0
        for i in range(df.shape[0]):
            if df.iloc[i,:].isnull()[5]:
                df.iloc[i,5] = fillnan_region_mwh_4(i/10000, i%10000/1000, i%1000/100, i%100/10, i%10, df,col,dat,n) # 填充缺失值方式
                if df.iloc[i,:].isnull()[5]:
                    count += 1
        print count
    return df
```
