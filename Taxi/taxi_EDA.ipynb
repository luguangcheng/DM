```python
from datetime import datetime
train_data['pickup_time'] = train_data['pickup_datetime'].map(lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M:%S'))
```

```python
train_data['pickup_time_month'] = train_data['pickup_time'].map(lambda x: x.month)
train_data.head()
```

```python
train_data['pickup_time_weekday'] = train_data['pickup_time'].map(lambda x: x.weekday())
train_data.head()
```

```python
train_data['pickup_time_hour'] = train_data['pickup_time'].map(lambda x: x.hour)
train_data.head()
```

```python
train_data['store_and_fwd_flag'] = train_data['store_and_fwd_flag'].map(lambda x: 0 if x=='N' else 1)
```

```python
train_data['weekday_sort'] = train_data['pickup_time_weekday'].map(c)
train_data.head()
```

```python
month_counts = train_data['pickup_time_month'].value_counts()
month_dict = dict(zip(list(month_counts.index.values), range(1,len(month_counts)+1)))
train_data['month_sort'] = train_data['pickup_time_month'].map(month_dict)
train_data.head()
```

```python
hour_counts = train_data['pickup_time_hour'].value_counts()
hour_dict = dict(zip(list(hour_counts.index.values), range(1,len(hour_counts)+1)))
train_data['hour_sort'] = train_data['pickup_time_hour'].map(hour_dict)
train_data.head()
```

```python
month_df = pd.get_dummies(train_data['pickup_time_month'], prefix='pickup_time_month')
week_df = pd.get_dummies(train_data['pickup_time_weekday'], prefix='pickup_time_weekday')
hour_df = pd.get_dummies(train_data['pickup_time_hour'], prefix='pickup_time_hour')
train_data = pd.concat([train_data,month_df,week_df,hour_df],axis=1)
train_data.head()
```

```python
train_data = pd.read_csv('train_data_0722_1.csv')
```

```python
train_data['pickup_longitude_cate'] = pd.qcut(train_data['pickup_longitude'],4,labels=range(4))
train_data.head()
```

```python
#删除经纬度异常数据
index = train_data[(train_data['dropoff_latitude']<40)|(train_data['dropoff_latitude']>41.5)|(train_data['pickup_latitude']<40)|(train_data['pickup_latitude']>41.5)|(train_data['dropoff_longitude']>-73)|(train_data['pickup_longitude']>-73)|(train_data['dropoff_longitude']<-75)|(train_data['pickup_longitude']<-75)].index
index
```

```python
train_data = train_data.drop(index)
train_data
```

```python
#经纬度离散化
n=5
train_data['pickup_longitude_qcut'] = pd.qcut(train_data['pickup_longitude'],n,labels=range(n))
train_data['dropoff_longitude_qcut'] = pd.qcut(train_data['dropoff_longitude'],n,labels=range(n))
train_data['dropoff_latitude_qcut'] = pd.qcut(train_data['dropoff_latitude'],n,labels=range(n))
train_data['pickup_latitude_qcut'] = pd.qcut(train_data['pickup_latitude'],n,labels=range(n))
train_data.head()
```

```python
#经纬度排序
n=10
train_data['pickup_longitude_cut'] = pd.cut(train_data['pickup_longitude'],n,labels=range(n))
train_data['dropoff_longitude_cut'] = pd.cut(train_data['dropoff_longitude'],n,labels=range(n))
train_data['dropoff_latitude_cut'] = pd.cut(train_data['dropoff_latitude'],n,labels=range(n))
train_data['pickup_latitude_cut'] = pd.cut(train_data['pickup_latitude'],n,labels=range(n))
train_data.head()
```

```python
def sort_feature(df,s):
    s_counts = df[s].value_counts()
    s_dict = dict(zip(list(s_counts.index.values), range(1,len(s_counts)+1)))
    df[s+'_sort'] = df[s].map(s_dict)
    return df
```

```python
#经纬度排序
train_data = sort_feature(train_data,'pickup_longitude_cut')
train_data = sort_feature(train_data,'dropoff_longitude_cut')
train_data = sort_feature(train_data,'dropoff_latitude_cut')
train_data = sort_feature(train_data,'pickup_latitude_cut')
train_data.head()
```
